## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

QUERY = """
SELECT *
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

# Create the histogram of ConvertedComp
plt.hist(df['ConvertedComp'])
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.title('Histogram of ConvertedComp')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Age'])
plt.title('Vertical Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
#Scatter plot
plt.scatter(df['Age'], df['WorkWeekHrs'])
plt.title('Scatter Plot of Age and WorkweekHrs')
plt.xlabel('Age')
plt.ylabel('WorkweekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Bubble plot
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

database_df = pd.read_sql_query(QUERY,conn)

# Count the occurrences of each database
database_counts = database_df['DatabaseDesireNextYear'].value_counts()

# Select top 5 databases
top_5_databases = database_counts.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show() 
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Filter data for the age group 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median values for 'WorkWeekHrs' and 'CodeRevHrs' for each age group
median_data = filtered_df.groupby('Age').median()

# Create a stacked chart using seaborn
plt.figure(figsize=(13, 6))
sns.barplot(x=median_data.index, y='WorkWeekHrs', data=median_data, color='blue', label='WorkweekHrs')
sns.barplot(x=median_data.index, y='CodeRevHrs', data=median_data, color='orange', label='CodeRevHrs', bottom=median_data['WorkWeekHrs'])

# Customize the plot
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend()

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Filter data for ages 45 to 60
# filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Filter data for ages 25 to 30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Calculate median 'ConvertedComp' for each age
median_data = filtered_df.groupby('Age')['ConvertedComp'].median().reset_index()

# Create a line chart using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='ConvertedComp', data=median_data, marker='o')

# Customize the plot
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp (Median)')
plt.grid(True)

# Show the plot
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Count the occurrences of each MainBranch
mainbranch_counts = df['MainBranch'].value_counts()

# Create a horizontal bar chart using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=mainbranch_counts, y=mainbranch_counts.index, orient='h')

# Customize the plot
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')

# Show the plot
plt.show()
```
